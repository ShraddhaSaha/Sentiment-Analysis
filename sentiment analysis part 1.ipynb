```python
columns = ['target', 'ids', 'date', 'flag', 'user', 'text']
data = pd.read_csv('training.1600000.processed.noemoticon.csv', encoding='ISO-8859-1', header=None, names=columns)
```

```python
data
```

```python
# Drop unnecessary columns
data = data.drop(['ids', 'date', 'flag', 'user'], axis=1)
```

```python
data
```

```python
# Convert target to human-readable labels
data['sentiment'] = data['target'].replace({0: 'Negative', 2: 'Neutral', 4: 'Positive'})
```

```python
data
```

```python
# Remove neutral tweets for this analysis
data = data[data['target'] != 2]
```

```python
data
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head(10))
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(data.describe())
```

```python
# Distribution of sentiment classes
print("\nDistribution of Sentiment Classes:")
print(data['sentiment'].value_counts())
```

```python
# Distribution of tweet lengths
data['text_len'] = data['text'].apply(len)
print("\nDistribution of Tweet Lengths:")
print(data['text_len'].describe())
```

```python
# Plotting distribution of tweet lengths
plt.figure(figsize=(10, 6))
sns.histplot(data['text_len'], bins=50, kde=True)
plt.title('Distribution of Tweet Lengths')
plt.xlabel('Tweet Length')
plt.ylabel('Frequency')
plt.show()
```

```python

# Word cloud of positive and negative tweets
from wordcloud import WordCloud
```

```python
# Concatenate positive and negative tweets
positive_tweets = ' '.join(data[data['sentiment'] == 'Positive']['text'])
negative_tweets = ' '.join(data[data['sentiment'] == 'Negative']['text'])
```

```python
# Generate word clouds
positive_wordcloud = WordCloud(width=800, height=400, background_color='white').generate(positive_tweets)
negative_wordcloud = WordCloud(width=800, height=400, background_color='white').generate(negative_tweets)
```

```python
# Plotting word clouds
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(positive_wordcloud, interpolation='bilinear')
plt.title('Positive Tweets Wordcloud')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(negative_wordcloud, interpolation='bilinear')
plt.title('Negative Tweets Wordcloud')
plt.axis('off')

plt.show()
```

```python
# Count the number of positive and negative sentiments
sentiment_counts = data['target'].value_counts()
```

```python
# Plotting Pie chart for sentiment distribution
plt.figure(figsize=(8, 6))
plt.pie(sentiment_counts, labels=['Negative', 'Positive'], autopct='%1.1f%%', startangle=140)
plt.title('Sentiment Distribution')
plt.axis('equal') 
plt.show()
```
